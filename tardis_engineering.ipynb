```python
# import data

import pandas as pd

phq = pd.read_csv("phq9_real_data.csv")
gad = pd.read_csv("gad7_real_data.csv")
```

```python
phq_mock = pd.read_csv("phq9_mock_data.csv")
gad_mock = pd.read_csv("gad7_mock_data.csv")

phq_mock
```

```python
# create asset

phq_asset = sy.Asset(
  name="TARDIS: PHQ Data",
  data = phq,
  mock = phq_mock
)

gad_asset = sy.Asset(
  name="TARDIS: GAD Data",
  data = gad,
  mock = gad_mock
)

phq_asset.data
```

```python
phq_asset.mock
```

```python
# create dataset

summary = "A validated outcome measurement dataset is a collection of..."

validated_outcome_measurement_dataset = sy.Dataset(
  name="TARDIS Validated Outcome Measurements",
  description="PHQ9 and GAD7",
  summary=summary
)
```

```python
validated_outcome_measurement_dataset
```

```python
validated_outcome_measurement_dataset.add_asset(phq_asset)
```

```python
validated_outcome_measurement_dataset
```

```python
validated_outcome_measurement_dataset.add_asset(gad_asset)
```

```python
client.upload_dataset(dataset=validated_outcome_measurement_dataset)
```
