```python
def print_attendance_outcomes(phq_data, attendance_col='attendance_rate'): 
  import pandas as pd
  phq_data['attendance_bucket'] = pd.cut(phq_data['attendance_rate'],
    bins=[0, 0.5, 0.75, 1.01],
    labels=['Low (<50%)', 'Medium (50-75%)', 'High (>75%)'])

  for bucket in ['Low (<50%)', 'Medium (50-75%)', 'High (>75%)']:
    rate = (phq_data[phq_data['attendance_bucket']==bucket]['outcome_label']=='Improved').mean()
    bar = '>' * int(rate*20)
    print (f" {bucket:<22} {bar} {rate:.0%}")

phq_mock = phq.mock
print_attendance_outcomes(phq_mock)
```
